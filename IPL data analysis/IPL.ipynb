```python
import numpy as np
import pandas as pd
```

```python
# Upload IPL Matches dataset into iplhistory dataframe
iplhistory = pd.read_csv('IPL Matches 2008-2020.csv')

# Upload IPLTeamNames dataset into iplteams dataframe
iplteams = pd.read_csv('IPL_Team_Names.csv')
```

```python
#copied iplhistory to ipl to perform all the required operations
ipl = iplhistory.copy()
```

```python
# Merge/join both dataframes to create a single iplhistory dataframe and replace team1, team2, winner columns with team names instead
a_list = list(iplteams['NickName'])
# print(a_list)
for i in range(len(a_list)):
 ipl.loc[ipl['team1'] == a_list[i],'team1'] = iplteams['TeamName'][i]
 ipl.loc[ipl['team2'] == a_list[i],'team2'] = iplteams['TeamName'][i]
 ipl.loc[ipl['winner'] == a_list[i],'winner'] = iplteams['TeamName'][i]
ipl.head(10)  
```

```python
# Update team1, team2 and winner column to reflect this name change           
# 						  Pune Warriors --> Rising Pune Supergaints
# 						  Delhi Daredevils --> Delhi Capitals
# 						  Deccan Chargers --> Sunrisers Hyderabad
# 						  Kings XI Punjab --> Punjab Kings
dxt = {'Pune Warriors':'Rising Pune Supergiants','Delhi Daredevils':'Delhi Capitals','Deccan Chargers':'Sunrisers Hyderabad','Kings XI Punjab':'Punjab Kings'}
for i in dxt.keys():
 ipl.loc[ipl['team1'] == i,'team1'] = dxt[i]
 ipl.loc[ipl['team2'] == i,'team2'] = dxt[i]
 ipl.loc[ipl['winner'] == i,'winner'] = dxt[i]

ipl.head(10) 
```

```python
# Update player names F Du Plessis --> Faf Du Plessis
ipl.loc[ipl['player_of_match'] == 'F du Plessis','player_of_match'] = 'Faf Du Plessis'
ipl['player_of_match'][754]
```

```python
#Delete rows where there was no winner i.e. abandoned games
ipl = ipl[ipl['winner'].notna()].reset_index(drop=True)
ipl['winner'][486]
```

```python
# Write a function to aggregate titles by franchise and display high to low
# since there was no information about titles or finals won, hence displayed most games won
def most_games_won(ipl):
    a = list(ipl['winner'].unique())
    dxt = {}
    for i in a:
        idx = list(ipl.index[ipl['winner'] == i])
        count = len(idx)
        dxt[i] = count

    sorted_dict = dict(sorted(dxt.items(), key= lambda item:item[1],reverse = True))
    for k,v in sorted_dict.items():
        print(k,":",v)

most_games_won(ipl)        
```

```python
# function to find the umpire to officiate in most IPL matches
def most_matches_umpire(ipl):
    u1 = list(ipl['umpire1'].unique())
    u2 = list(ipl['umpire2'].unique())
    ump1 = {}
    for i in u1:
        idx = list(ipl.index[ipl['umpire1'] == i])
        count = len(idx)
        ump1[i] = count

    for i in u2:
        idx = list(ipl.index[ipl['umpire2'] == i])
        count = len(idx)
        if i in ump1.keys():
          ump1[i]+= count
        else:
            ump1[i] = count    

    sorted_ump1 = dict(sorted(ump1.items(), key= lambda item:item[1],reverse = True))
    print(list(sorted_ump1.keys())[0])

most_matches_umpire(ipl)        
```

```python
# function to find all home teams that won eliminator games batting first
def home_eliminator_won(ipl):
    idx = list(ipl.index[ipl['eliminator'] == 'Y'])
    for i in idx:
        if ipl['winner'][i] == ipl['team1'][i]:
            print(ipl['winner'][i])  

home_eliminator_won(ipl)
```

```python
# function to take list of columns as input and Drops columns from the dataframe permanently (METHOD and ID columns)
def drop_columns(ipl,col):
  df = ipl.drop(columns= col)
  return df
l = ['id','method']
ipl = drop_columns(ipl,l)  
ipl.head()
```

```python
# function to calculate avg win margin for teams batting first at Chennai
def avg_win_margin_chennai(ipl):
    df = ipl.loc[(ipl['city'] == 'Chennai') & (ipl['toss_decision'] == 'bat') & (ipl['result'] == 'runs')]
    lt = list(df['result_margin'])
    return(sum(lt)/len(lt))

print(avg_win_margin_chennai(ipl))
```

```python
# First,separating year in date column and creating a new column for it
ipl['year'] = ipl['date'].str.split('/').str[2]
ipl.head()
```

```python
# Which team participated in the least number of IPL seasons, how many?
team = ipl['team1'].unique()
yr = ipl['year'].unique()
season = {}
for i in yr:
    team_participate = {}
    for j in team:
        idx = ipl.index[(ipl['team1'] == j) & (ipl['year']==i)]
        count = len(idx)
        if(count > 0):
            team_participate[j] = 1
            season[i] = team_participate

participation = {}
for t in team:
    participation[t] = 0
# print(season)
for k1,v1 in season.items(): 

 for k2,v2 in v1.items():
    if participation[k2]:
       participation[k2]+=v2
    else:
       participation[k2] = v2

min_val = min(participation.values())
for k,v in participation.items():
   if v == min_val:
      print(k,":",v)                  
```

```python
# player who won the most man of the match awards per year, along with the number of awards
player = ipl['player_of_match'].unique()
yr = ipl['year'].unique()
season = {}
for i in yr:
    man_of_match = {}
    for j in player:
        idx = ipl.index[(ipl['player_of_match'] == j) & (ipl['year']==i)]
        count = len(idx)
        if(count > 0):
            man_of_match[j] = count
            season[i] = man_of_match

for k in season.keys(): 
 sorted_mom = dict(sorted(season[k].items(), key= lambda item : item[1],reverse=True))
 max_val = max(sorted_mom.values())
 for key,v in sorted_mom.items():
    if v == max_val:
       print(k,"-> {}".format(key),"-> {}".format(v))        
```

```python
# team/list of teams which were involved in the max number of tied matches across all seasons
yr = ipl['year'].unique()
team = ipl['team1'].unique()
for i in yr:
    print("In: {}".format(i))
    team_tie = {}
    for t in team:
       idx = ipl.index[(ipl['year'] == i) & (ipl['team1'] == t) & (ipl['result'] == 'tie')]
       count = len(idx)
       team_tie[t] = count
    for t in team:
       idx = ipl.index[(ipl['year'] == i) & (ipl['team2'] == t) & (ipl['result'] == 'tie')]
       count = len(idx)
       if team_tie[t]:
         team_tie[t]+= count 
       else: 
         team_tie[t] = count
    max_val = max(team_tie.values()) 
    for k,v in team_tie.items():
       if v == max_val:
            print(k,":",v)   
    print("\n")   
```

```python
# # number of matches won by the teams batting second for each venue, for each season
yr = ipl['year'].unique()
ven = ipl['venue'].unique()
team = ipl['team1'].unique()
season = {}

for i in yr:
    venue = {}
    for v in ven:
       matches_won = {}
       for t in team:
           idx = ipl.index[(ipl['year'] == i) & (ipl['venue'] == v) & (ipl['winner'] == t) & (ipl['result'] == 'wickets')]
           count = len(idx)
           matches_won[t] = count
       venue[v] = matches_won
    season[i] = venue

for k1,v1 in season.items():
    for k2,v2 in v1.items():
        for k3,v3 in v2.items():
            print(k1,"->",k2,"->",k3,"->",v3)
```
