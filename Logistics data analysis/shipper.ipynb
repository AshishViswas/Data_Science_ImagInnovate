```python
import numpy as np
import pandas as pd
from faker import Faker
```

```python
data = pd.read_csv('internsdata_updated.csv')
```

```python
# To find No.of Unique Account Id
account_id = list(data['account_id'].unique())
print("No.of unique account id's: {}".format(len(account_id)))
```

```python
# Execute only once
# it generates fake account names and account codes for every distinct account id
fake = Faker(['en_US'])
for i in account_id:
    a = fake.name()
    data.loc[data['account_id'] == i, 'account_name'] = a
    list = a.split(" ")
    c =""
    for j in list:
      c = c + j[0]
    data.loc[data['account_id'] == i, 'account_code'] = c  
```

```python
data.head()
```

```python
# grouping values based on account_name
data.sort_values(by = ['account_name'], inplace = True)
data.reset_index(inplace=True)
```

```python
# to see no.of NaN values present in each column
data.isnull().sum()
```

```python
# Importing random and datetime modules to manipulate datetime fields
import random
from datetime import datetime, timedelta
```

```python
def generate_random_offset():
    return random.randint(3,30) # 30 days
```

```python
# Replace all the NaN values from the actual start date column with the created time
data['actual_start_datetime'] = data['actual_start_datetime'].fillna(data['created_at'])
```

```python
# to perform datetime operations on a column, we first need to convert it to datetime type
# Convert 'actual_start_datetime' column to datetime
data['actual_start_datetime'] = pd.to_datetime(data['actual_start_datetime'])
```

```python
# for every company assign an appointed time with maxx of 30 days but each company should have something close duration
# Creating new column 'appointment_at' by assigning a random time value in 30 day interval on 'actual_start_datetime'
#Execute only once since time values change randomly on every execution
data['appointment_at'] = data['actual_start_datetime'] - data['actual_start_datetime'].apply(lambda x: timedelta(seconds=generate_random_offset()))
```

```python
data.head()
```

```python
# dropping null values in 'account_id' column so as to ignore unknown accounts and as there are only 1097 records which is just 0.38 % of total data
data.dropna(subset = ['account_id'],inplace = True)
```

```python
df = data.copy()
```

```python
# Calculate new time values for each company
company_offsets = {}
for company, group in df.groupby('account_name'):
    offset = generate_random_offset()  # Your existing function
    company_offsets[company] = offset

df['appointment_at'] = df.apply(lambda row: row['actual_start_datetime'] - timedelta(days = company_offsets[row['account_name']] + np.random.randint(-3,3)), axis=1)
```

```python
# Calculate time difference in seconds
df['time_difference'] = (df['actual_start_datetime'] - df['appointment_at']).dt.total_seconds()

# Create 'expedite_level' column and assign 'expedite_sharp' to those whose time difference is less than 24 hours and just 'expedite' to others 
df['expedite_level'] = df['time_difference'].apply(lambda diff: 'expedite_sharp' if diff < 24 * 60 * 60 else 'expedite')
```

```python
df.head()
```

```python
# to see no.of orders which are expedite or expedite_sharp
df['expedite_level'].value_counts()
```

```python
df['Lead_Time'] = df['actual_start_datetime'] - df['appointment_at'] 
```

```python
#Dropping 'height', 'length', 'width', 'scheduled_start_datetime', 'scheduled_end_datetime', 'account_code', 'account_id' and 'time_difference' columns as they are unneccessary and not considered while reporting
df.drop(columns = ['account_id', 'height', 'length', 'width','scheduled_start_datetime',
        'scheduled_end_datetime', 'account_code','time_difference'],inplace = True) 
```

```python
#also dropping the 'index' and 'Unnamed: 0' and 'customer_order_id' columns
df.drop(columns = ['index','Unnamed: 0','customer_order_id'],inplace = True)
```
